```python
import time


class Timer(object):
    def __init__(self, verbose=False):
        self.verbose = verbose

    def __enter__(self):
        self.start = time.time()
        return self

    def __exit__(self, *args):
        self.end = time.time()
        self.secs = self.end - self.start
        self.msecs = self.secs * 1000  # millisecs
        if self.verbose:
            print('elapsed time: %f ms' % self.msecs)
        print('elapsed time: %f s' % self.secs)
```

```python
"""
By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.
What is the 10 001st prime number?
"""


def find_prime_numbers(limit):
    prime_lst = []
    for i in range(2, 10**1000):
        if len(prime_lst) == limit:
            break
        for y in range(2, 10**1000):
            if i % y == 0:
                if y not in prime_lst:
                    prime_lst.append(y)
                break
    #print(prime_lst[limit-1])
    return prime_lst[-1]


number = 1000
with Timer() as p:
    print(find_prime_numbers(number))
```

```python
def nextIsPrime(number):
    for i in range(2, int(number*number)):
        if number % i ==0:
            return False
    return True


def atlernative_prime(prime_limit):
    lst = [2]
    next_prime = 3
    while len(lst) < prime_limit:
        if nextIsPrime(next_prime):
            lst.append(next_prime)
        next_prime +=1
    return prime_lst[-1]

print(atlernative_prime(100))
```

```python
#!/usr/bin/env python
import math
import threading
import pprofile
from time import time
import sys


def is_prime(num):
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True


def get_prime_numbers(count):
    prime_numbers = [2]
    next_number = 3

    while len(prime_numbers) < count:

        if is_prime(next_number):
            prime_numbers.append(next_number)
        next_number += 1
        
    return prime_numbers[-1]


#Single-treaded run
prof = pprofile.Profile()
with prof:
    print(get_prime_numbers(1000))
prof.annotate(sys.stdout, __file__)
```
