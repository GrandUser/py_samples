```python
import requests
import pandas as pd
from sys import argv       
from datetime import datetime

        
def create_links():
    args_lst = []
    for arg in argv:
        args_lst.append(arg)
    del args_lst[0] # del file name arg
    try:
        start_date = str(args_lst[1]) if len(args_lst)>1 else '19521015'
        end_date = str(args_lst[2]) if len(args_lst)>2 else datetime.today().strftime('%Y%m%d')
        base_link = 'https://www.stoxx.com/itf/fqs/delayed/charts.json?'
        base_params = 'select=ValorId,ValorSymbol,ISIN,LastPrice,DailyChange&'
        link = base_link + base_params + 'where=ValorSymbol={}&columns=Millis,Date,Close,Open,High,Low&netting=1440&fromdate={}&todate={}'.format(str(args_lst[0]), start_date, end_date)
        print(link)
    except:
        print(link)
        print('wrong link')
    return link

def get_raw_data_from_STOXX(_link=argv[1]):
    raw_data = requests.get(_link)
    data = raw_data.json()['valors'][0]
    return data

def get_ISIN(_data):
    return data['ISIN']
    
def get_dataframe_from_raw_data(_data):
    ohlc_data = data['data']
    df = pd.DataFrame(ohlc_data, index=ohlc_data['Date'], columns=['Open', 'High', 'Low', 'Close', 'Millis'])
    return df

def save_data_to_excel(_dataframe, _ISIN):
    with open('{}.csv'.format(_ISIN), "a") as f:
        try:
            _dataframe.to_csv(f, header=True)
        except Exception as e:
            print(e)
    print('saved to {}'.format(_ISIN))
            
if __name__ == "__main__":
    link = 'https://www.stoxx.com/itf/fqs/delayed/charts.json?select=ValorId,ValorSymbol,ISIN,LastPrice,DailyChange&where=ValorSymbol=BKXP&columns=Millis,Date,Close,Open,High,Low&netting=1440&fromdate=19911015&todate=20160408'
    #link = create_links()
    data = get_raw_data_from_STOXX(link)
    ISIN = get_ISIN(data)
    df = get_dataframe_from_raw_data(data)
    save_data_to_excel(df, ISIN)
```
